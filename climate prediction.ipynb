```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Load the CSV files (Ensure that the paths are correct in your environment)
train_df = pd.read_csv('/content/DailyDelhiClimateTrain.csv')
test_df = pd.read_csv('/content/DailyDelhiClimateTest.csv')

# Drop 'date' and 'meantemp' columns from the training data
X_train = train_df.drop(columns=['meantemp', 'date'])  # Features: humidity, wind_speed, meanpressure
y_train = train_df['meantemp']  # Target: meantemp (temperature)

# Drop 'date' and 'meantemp' from the test data (no target in the test data)
X_test = test_df.drop(columns=['date', 'meantemp'])  # Same features used in training

# Train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the test data
predictions = model.predict(X_test)

# Display predictions
print("Predictions on Test Data:")
print(predictions)

# Optional: Visualizing the predictions
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
plt.plot(predictions, label='Predicted Temperature')
plt.title('Predicted Temperature on Test Data')
plt.xlabel('Sample')
plt.ylabel('Temperature (C)')
plt.legend()
plt.show()
```
